```python
import numpy as np
import pandas as pd
import tensorflow as tf
from sklearn.preprocessing import LabelEncoder, OneHotEncoder, StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, accuracy_score
```

```python
# Load Dataset
dataset = pd.read_csv('Churn_Modelling.csv')
X = dataset.iloc[:, 3:-1].values # Select features
y = dataset.iloc[:, -1].values # Target variable #
```

```python
# Encoding Categorical Data
le = LabelEncoder()
X[:, 2] = le.fit_transform(X[:, 2]) # Label Encoding "Gender"
```

```python
print(X)
```

```python
# One-Hot Encoding "Geography"
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [1])], remainder='passthrough')
```

```python
X = np.array(ct.fit_transform(X))
```

```python
# Splitting the Dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
# Feature Scaling
sc = StandardScaler()
X_train[:, 3:] = sc.fit_transform(X_train[:, 3:])
X_test[:, 3:] = sc.transform(X_test[:, 3:])
```

```python
ann.summary()
```

```python
# Compiling & Training the ANN #for gradient descent
ann.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
ann.fit(X_train.astype(int), y_train, batch_size=32, epochs=100)
```

```python
# Making Predictions
y_pred = ann.predict(X_test.astype(int))
y_pred = (y_pred > 0.5) #threshold
```

```python
# Evaluating Performance
cm = confusion_matrix(y_test, y_pred)
accuracy = accuracy_score(y_test, y_pred)
```
